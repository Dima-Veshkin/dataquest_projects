```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style as style
from datetime import datetime as dt
%matplotlib inline
style.use('fivethirtyeight')


exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
exchange_rates.info()
```

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)


exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
exchange_rates.head()
```

```python
euro_to_dollar = exchange_rates[['Time','US_dollar']].copy()
euro_to_dollar['US_dollar'].value_counts()
```

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
euro_to_dollar.info()
```

```python
plt.plot(euro_to_dollar['Time'],euro_to_dollar['US_dollar'])
plt.title('EURO-USD rate average 1999-2021')
plt.show()
```

```python
euro_to_dollar['roll_mean_30'] = euro_to_dollar['US_dollar'].rolling(30).mean()

plt.figure(figsize=(15,7))
plt.plot(euro_to_dollar['Time'],euro_to_dollar['roll_mean_30'])
plt.title('EURO-USD rate average 1999-2021')
plt.show()
```

Here are a few story ideas for our data:

- We show how the euro-dollar rate has changed during the coronavirus pandemic. We can show the 2020 data and the 2016-2019 data as a baseline. We can use a line plot.
- We show how the euro-dollar rate changed during the 2007-2008's financial crisis. We can also show the data for 2016 and 2009 for comparison. We can use a line plot.
- We show comparatively how the euro-dollar rate changed under the last three US presidents (George W. Bush (2001-2009), Barack Obama (2009-2017), and Donald Trump (2017-2021)). We can use a line plot.

<b>I will choose the history of the financial crisis of 2007-2008. I am interested in this story, also thanks to my favorite films 'Too Big to Fail' and 'The Big Short'</b>

## Financial Crisis

```python
fin_crisis_all = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2006) & 
                                       (euro_to_dollar['Time'].dt.year <= 2009)]
fin_crisis_peak = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2007) & 
                                       (euro_to_dollar['Time'].dt.year <= 2008)]
```

```python
fig, ax = plt.subplots(figsize=(15,7))

# задаем фон всему
#fig.patch.set_facecolor('#f0f0f0') 

plt.plot(fin_crisis_all['Time'],fin_crisis_all['roll_mean_30'], color='#A6D785', alpha=0.5)
plt.plot(fin_crisis_peak['Time'],fin_crisis_peak['roll_mean_30'], color='#e23d28')

# определяем тики на х, по поиску года среди дат
ax.set_xticks([dt.strptime('2006', '%Y'),dt.strptime('2007', '%Y'),
               dt.strptime('2008', '%Y'),dt.strptime('2009', '%Y'),
               dt.strptime('2010', '%Y')])
ax.set_xticklabels([2006,2007,2008,2009,2010], size=15, c='grey')

# определяем тики на y, и убираем первый тик для визуальной красоты
ax.set_yticks([1.6,1.5,1.4,1.3,1.2])
ax.set_yticklabels(["",1.5,1.4,1.3,1.2], size=15, c='grey')

# рисуем тест, главное выбрать правильную х координату
ax.text(732192.0, 1.115, '©Dima Veshkin' + ' '*160+'Source: European Central Bank',
        color = '#ffffff', backgroundcolor = '#4d4d4d',
        size=13)
ax.text(732192.0, 1.65,"Euro-USD rate peaked at 1.59 during 2007-2008's financial crisis",
        c='black',size=24,weight='bold')
ax.text(732192.0, 1.62,"Euro-USD exchange rates between 2006 and 2010",
        c='black',size=19)

# выделяем пик
ax.axvspan(xmin=733112.0, xmax=733302.0, ymin=0.09,
           alpha=0.3, color='grey')
# делаем сетку прозрачнее
ax.grid(alpha=0.5)

plt.show()
```

## The Three US Presidencies


```python
bush_obama_trump = euro_to_dollar.copy(
                   )[(euro_to_dollar['Time'].dt.year >= 2001) & (euro_to_dollar['Time'].dt.year < 2021)]
bush = bush_obama_trump.copy(
       )[bush_obama_trump['Time'].dt.year < 2009]
obama = bush_obama_trump.copy(
       )[(bush_obama_trump['Time'].dt.year >= 2009) & (bush_obama_trump['Time'].dt.year < 2017)]
trump = bush_obama_trump.copy(
       )[(bush_obama_trump['Time'].dt.year >= 2017) & (bush_obama_trump['Time'].dt.year < 2021)]
```

```python
plt.figure(figsize=(12,6))
ax1 = plt.subplot(2,3,1)
ax2 = plt.subplot(2,3,2)
ax3 = plt.subplot(2,3,3)
ax4 = plt.subplot(2,1,2)
axes = [ax1, ax2, ax3, ax4]

ax.set_yticks([1.6,1.5,1.4,1.3,1.2])
ax.set_yticklabels(["",1.5,1.4,1.3,1.2], size=15, c='grey')

for ax in axes:
    ax.set_ylim(0.8,1.7)
    ax.set_yticks([1.0,1.2,1.4,1.6])
    ax.set_yticklabels([1.0,1.2,1.4,1.6] ,alpha = 0.4)
    ax.grid(alpha=0.5)
    
ax1.plot(bush['Time'],bush['roll_mean_30'],color='#BF5FFF')
ax1.set_xticklabels([2001,'',2003,'',2005,'',2007,'',2009] ,alpha = 0.4)
ax1.text(731516.0, 1.92, 'BUSH', fontsize=18, weight='bold',
        color='#BF5FFF')
ax1.text(731416.0, 1.8, '(2001-2009)', weight='bold',
        alpha=0.3)   

ax2.plot(obama['Time'],obama['roll_mean_30'],color='#ffa500')
ax2.set_xticklabels([2009,'',2011,'',2013,'',2015,'',2017] ,alpha = 0.4)
ax2.text(734316.0, 1.92, 'OBAMA', fontsize=18, weight='bold',
        color='#ffa500')
ax2.text(734366.0, 1.8, '(2009-2017)', weight='bold',
        alpha=0.3)  

ax3.plot(trump['Time'],trump['roll_mean_30'],color='#00B2EE')
ax3.set_xticklabels([2017,'',2018,'',2019,'',2020,'',2021] ,alpha = 0.4)
ax3.text(736816.0, 1.92, 'TRUMP', fontsize=18, weight='bold',
        color='#00B2EE')
ax3.text(736816.0, 1.8, '(2017-2021)', weight='bold',
        alpha=0.3)

ax4.plot(bush['Time'],bush['roll_mean_30'],color='#BF5FFF')
ax4.plot(obama['Time'],obama['roll_mean_30'],color='#ffa500')
ax4.plot(trump['Time'],trump['roll_mean_30'],color='#00B2EE')
ax4.grid(alpha=0.5)
ax4.set_xticks([])

### Adding a title and a subtitle
ax1.text(730016.0, 2.35, 'EURO-USD rate averaged 1.22 under the last three US presidents',
         fontsize=20, weight='bold')
ax1.text(730016.0, 2.14, '''EURO-USD exchange rates under George W. Bush (2001 - 2009), Barack Obama (2009-2017),
and Donald Trump (2017-2021)''',
        fontsize=16)

### Adding a signature
ax4.text(729806.0, 0.65, '©DATAQUEST' + ' '*106 + 'Source: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=14)

plt.show()
```
